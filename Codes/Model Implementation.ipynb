```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from joblib import dump
```

```python
df=pd.read_csv('Data.csv')
df
```

```python
df = df[df['city'] =='Lahore'].copy().reset_index(drop=True)
df.shape
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.drop(['agency', 'agent'], axis=1,inplace=True)
df.isnull().sum()
```

```python
df.drop(['Unnamed: 0', 'property_id','location_id','page_url','purpose','date_added','city','province_name','location'], axis=1,inplace=True)
```

```python
df.drop_duplicates(inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
df.property_type.value_counts()
```

```python
df.property_type.unique()
```

```python
df.groupby('property_type')['price'].mean().reset_index().sort_values(by='price',ascending=True)
```

```python
#rank 0 to "Upper Portion",likewise rank 6 to "Farm House"
from pandas.api.types import CategoricalDtype
property_mapping = ["Upper Portion", "Lower Portion", "Flat", "Room", "Penthouse", "House", "Farm House"]
property_dtype = pd.CategoricalDtype( categories=property_mapping,ordered=True)
df['PropertyType'] = df['property_type'].astype(property_dtype).cat.codes
```

```python
df['PropertyType'].unique()
```

```python
df.drop('property_type',axis=1,inplace=True)
df
```

# Outlier Removed

```python
from scipy import stats
import pandas as pd
z_scores = stats.zscore(df)
threshold = 7
outliers_mask = (abs(z_scores) > threshold).any(axis=1)
df_cleaned = df[~outliers_mask]
df_cleaned.reset_index(drop=True, inplace=True)
df_cleaned.to_csv("OutliersRemoved.CSV",index=False)
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
standardized_data = scaler.fit_transform(df)
standardized_df = pd.DataFrame(standardized_data, columns=df.columns)
standardized_df.to_csv("standardized.CSV",index=False)
```

# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
feature_scaler = MinMaxScaler()
price_scaler = MinMaxScaler()
normalized_features = feature_scaler.fit_transform(df_cleaned.drop('price', axis=1))
price_scaler.fit(df_cleaned[['price']])
dump(price_scaler, 'price_scaler.joblib')
dump(feature_scaler, 'feature_scaler.joblib')
normalized_df = pd.DataFrame(normalized_features, columns=df_cleaned.drop('price', axis=1).columns)
normalized_df['price'] = price_scaler.fit_transform(df_cleaned[['price']])
normalized_df.to_csv("normalized.csv", index=False)
normalized_df
```

# Model training testing

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
#X = standardized_df.drop('price', axis=1)
#y = standardized_df['price']  
X = normalized_df.drop('price', axis=1)
y = normalized_df['price'] 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=41)
Normalized_model= RandomForestRegressor(n_estimators=300,max_depth=20,random_state=41)
Normalized_model.fit(X_train, y_train)
y_pred = Normalized_model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print(f"R2 Score: {r2}")
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
```

```python
from joblib import dump
from joblib import load
dump(Normalized_model, 'Normalized_model.joblib')
```

```python
loaded_model = load('Normalized_model.joblib')
```

```python
new_instance = {
    "latitude": 31.499348, 
    "longitude": 74.416959, 
    "baths": 6, 
    "bedrooms": 3, 
    "Total_Area": 2450.259,
    "PropertyType": 5
}
feature_scaler = load('feature_scaler.joblib')
new_instance_df = pd.DataFrame([new_instance])
new_instance_normalized=feature_scaler.fit_transform(new_instance_df)
predicted_price =loaded_model.predict(new_instance_normalized)[0]
print(f"The predicted price is: {predicted_price}")
```

```python
price_scaler = load('price_scaler.joblib')
predicted_normalized_price = np.array([predicted_price])
predicted_normalized_price = predicted_normalized_price.reshape(-1, 1)
predicted_price_original_scale = price_scaler.inverse_transform(predicted_normalized_price)
predicted_price2 = predicted_price_original_scale[0, 0]
print(f"The predicted price in original scale is: {predicted_price2}")
```

# HYPER TUNNING

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(estimator=RandomForestRegressor(random_state=42), 
                           param_grid=param_grid, 
                           cv=3, 
                           n_jobs=-1, 
                           verbose=2)
grid_search.fit(X_train, y_train)
print("Best parameters:", grid_search.best_params_)
best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred_best)
mse = mean_squared_error(y_test, y_pred_best)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred_best)
print(f"R2 Score: {r2}")
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
```
